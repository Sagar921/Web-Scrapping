# Write a python program to scrape ‘software developer’ job listings from ‘Monster.com’. It should include all the jobs listed for the next 5 pages. 

```python
# Import essential libraries

from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Configure webdriver to use Chrome Browser

driver = webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://www.monsterindia.com/srp/results?query=software%20developer&locations=Mumbai,Mumbai%20City,Navi%20Mumbai,Thane&searchId=4240e019-af90-45cc-b730-a7181be79c72")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
print(soup)
```

```python
location = soup.find_all('div', class_ = 'col-xxs-12 col-sm-5 text-ellipsis')

Location = []

for i in location:
    Location.append(i.text)
    
    
Newlocation=[]
for i in Location:
    newname=i.replace("\n","")
    Newlocation.append(newname)
    
len(Newlocation)
```

```python
years = soup.find_all('div', 'exp col-xxs-12 col-sm-3 text-ellipsis')

Experience = []

for i in years:
    Experience.append(i.text)
    
    
Newexperience=[]
for i in Experience:
    newname=i.replace("\n","")
    Newexperience.append(newname)
    
len(Newexperience)
```

```python
salary = soup.find_all('div', 'package col-xxs-12 col-sm-4 text-ellipsis')

Salary = []

for i in salary:
    Salary.append(i.text)
    
    
Newsalary=[]
for i in Salary:
    newname=i.replace("\n","")
    Newsalary.append(newname)

len(Newsalary)
```

```python
skill = soup.find_all('p', 'descrip-skills')

Skills = []

for i in skill:
    Skills.append(i.a.text)
    
    
Newskills=[]
for i in Skills:
    newname=i.replace("\n","")
    Newskills.append(newname)
    
len(Newskills)
```

```python
company = soup.find_all('span', class_ = 'company-name')

Company_Name = []

for i in company:
    Company_Name.append(i.text)
    
    
Newcompany=[]
for i in Company_Name:
    newname=i.replace("\n","")
    Newcompany.append(newname)
    
len(Newcompany)
```

```python
pg_1 = pd.DataFrame({'Company_Name': Newcompany, 'Location': Newlocation, 'Experience': Newexperience, 'Salary':Newsalary,'Skills':Newskills})
pg_1
```
