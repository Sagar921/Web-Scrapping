# 2)Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver= webdriver.Chrome("chromedriver.exe")
```

```python
products=[]
saal=[]
rate=[]
driver.get("https://www.imdb.com/chart/top?sort=ir,desc&mode=simple&page=1")
```

```python
content= driver.page_source
soup= BeautifulSoup(content)
names=soup.findAll("td", attrs={"class":"titleColumn"})
years=soup.findAll("span", attrs={"class":"secondaryInfo"})
rating=soup.findAll("td", attrs={"class":"ratingColumn imdbRating"})
print(len(names), len(years), len(rating))
```

```python
print(len(names))
```

```python
for name in names:
    products.append(name.text)
```

```python
products
```

```python
newproducts=[]
for i in products:
    prd=i.replace("\n","")
    newproducts.append(prd)
```

```python
newproducts
```

```python
#print(products[0])
#len(products)
```

```python
for year in years:
    saal.append(year.text)
```

```python
saal
```

```python
newsaal=[]
for i in saal:
    newsal=i.replace("\n","")
    newsaal.append(newsal)
```

```python
newsaal
```

```python
#print(saal[0])
#len(saal)
```

```python
for Ratings in rating:
    rate.append(Ratings.text)
```

```python
rate
```

```python
newrating=[]
for i in rate:
    newrate=i.replace("\n","")
    newrating.append(newrate)
```

```python
newrating
```

```python
#print(rate[0])
#len(rate)
```

```python
df=pd.DataFrame({"Names":newproducts,"Years":newsaal, "Rating":newrating})
```

```python
df.to_csv("picture.csv", index=False, encoding="utf-8-sig")
```
