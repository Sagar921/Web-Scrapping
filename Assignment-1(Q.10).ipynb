# Write a python program to scrape ‘data scientist’ job listings for location ‘New Delhi’ from ‘Monster.com’. It should include all the jobs listed for the next 5 pages.

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
driver= webdriver.Chrome("chromedriver.exe")
```

```python
url="https://www.monsterindia.com/srp/results?query=Data%20Scientist&locations=Delhi&experienceRanges=0~0&experience=0&searchId=8d6758b6-12d4-4ff5-8138-4edd459f8c40"
request=requests.get(url)
request
```

```python
driver.get(url)
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
print(soup)
```

```python
location = soup.find_all('div', class_ = 'col-xxs-12 col-sm-5 text-ellipsis')

Location = []

for i in location:
    Location.append(i.text)
    
    
Newlocation=[]
for i in Location:
    newname=i.replace("\n","")
    Newlocation.append(newname)
    
len(Newlocation)
```

```python
years = soup.find_all('div', 'exp col-xxs-12 col-sm-3 text-ellipsis')

Experience = []

for i in years:
    Experience.append(i.text)
    
    
Newexperience=[]
for i in Experience:
    newname=i.replace("\n","")
    Newexperience.append(newname)
    
len(Newexperience)
```

```python
jobtitle = soup.find_all('h3', 'medium')

Jobtitle = []

for i in jobtitle:
    Jobtitle.append(i.text)
    
    
Newjobtitle=[]
for i in Jobtitle:
    newname=i.replace("\n","")
    Newjobtitle.append(newname)

len(Newjobtitle)
```

```python
pg_1 = pd.DataFrame({'Job_Title': Newjobtitle, 'experiance': Newexperience, 'location':Newlocation})
pg_1
```
